1D NumPy Array – Element-wise Operations

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([2, 4, 5])


#Addition
print(f"1D NumPy Array – Addition: {a+b}")

#Subtraction
print(f"1D NumPy Array – Subtraction: {a-b}")

#Multiplication
print(f"1D NumPy Array –Multiplication: {a*b}")

#Division
print(f"1D NumPy Array – Division: {a/b}");
```

2D NumPy Array – Element-wise Operations

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

#Addition
print(f"2D NumPy Array – Addition: {A+B}")

#Subtraction
print(f"2D NumPy Array – Subtraction: {A-B}")

#Multiplication
print(f"2D NumPy Array –Multiplication: {A*B}")

#Division
print(f"2D NumPy Array – Division: {A/B}");
```

Dot product in Numpy & Pytorch

```python
#Numpy array dot product using np.dot()
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.dot(a, b)
print(result)


#pytorch tensor dot product using torch.dot()
x = torch.tensor([10, 20, 30])
y = torch.tensor([2, 4, 5])
result = torch.dot(x, y)
print(result)

```

Matrix Multiplication for 2D Numpy aaray and pytorch tensor

```python
A = np.array([[1, 2],
                  [3, 4]])

B = np.array([[5, 6],
                  [7, 8]])

print(f"2D Numpy array-matrix multiplication: {A@B}");


A = torch.tensor([[1, 2],
                  [3, 4]])

B = torch.tensor([[5, 6],
                  [7, 8]])

print(f"2D pytorch tensor -matrix multiplication: {A@B}");

```

Indexing and Slicing examples (Boolean masking, extracting subtensors)

```python
import torch
import numpy as np

t = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])
#indexing
print(t[0, 1])

#slicing
print(t[:,1])
print(t[0,:])

#extracting subtensor in tensor
print(t[0:2,1:2])


```

```python
#boolean masking: Selecting elements from an array using a condition that returns True or False.
#boolean masking on tensor
mask = t > 30
print(mask)
print(t[mask])

#boolean masking on array
arr=np.array([[10,40,30],[20,65,26],[56,12,34]]);
mask=arr>30
print(mask)
print(arr[mask])

#extracting submatrix in array
print(arr[0:2,1:2])


```

```python
# .view()
import torch
x=torch.tensor([1,2,3,4,5,6])
y=x.view(2,3)
print(y)
print(x)

# .reshape()
z=x.reshape(2,3)
print(z)
print(x)

#.unsqueeze()
y = x.unsqueeze(0)
print(y)
print(y.shape)

z=x.unsqueeze(1)
print(z)
print(z.shape)


#.squeeze()
print(y.squeeze(0))
print(z.squeeze(1))



```

Broadcasting- Perform operations with tensors of different shapes.

```python
import torch

a = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

b = torch.tensor([10, 20, 30])

print(a + b)
```

```python
#out-of-place operation
x = torch.tensor([1, 2, 3])
y = x + 5

print(x)
print(y)


#in-place operation
x = torch.tensor([1, 2, 3])
x.add_(5)

print(x)

```
